```python
import pandas as pd
import numpy as np


train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

## Clean

```python
train.replace(-999, np.nan, inplace=True)
```

```python
to_drop = train.columns[(train.isna().sum() >= int(0.5*train.shape[0])).values]
train.drop(to_drop, axis=1, inplace=True)
train.replace(np.nan, train.median(), inplace = True)
```

```python
train["Prediction"] = np.where(train["Prediction"]=="s", 1, 0)
```

```python
train.shape
```

## Pycaret tests

```python
from pycaret.classification import *
```

Simple pycaret test with all defaults... https://pycaret.org/classification/.
For logistic regression, we do not need to use the full sample to train the ML model and still get good accuracy in K-Fold CV.  Selecting sample to 0.2, speeds up the computations and does not affect the predictions.

```python
clf = setup(data=train, target = "Prediction")
```

```python
lr = tune_model("lr", n_iter=50)
```

```python
evaluate_model(lr)
```

## Try with all the bells and whistles.

```python
data_subset = train[int(0.95*train.shape[0]):]
data_subset
```

```python
# Takes a while
clf2 = setup(data=data_subset, target = "Prediction", normalize = True, normalize_method="zscore",
            polynomial_features = True, polynomial_degree = 4, feature_interaction=True, feature_ratio=True)
```

```python
lr2 = tune_model("lr")
```

```python
evaluate_model(lr2)
```

```python
# Train the tuned model on the entire training set now.
model = finalize_model(lr2)
```

Using the trained model, check the test accuracy

```python
# Clean
test.replace(-999, np.nan, inplace=True)
test.drop(to_drop, axis=1, inplace=True)
test.replace(np.nan, train.median(), inplace = True)
```

```python
# Takes a while because it needs to recompute again all of the feature interactions & ratios and polynomials etc... 
predictions = predict_model(model, data=test[int(0.98*test.shape[0]):])["Label"]
```
